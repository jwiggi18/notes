# Dataset

**Historical Wildfires**

This wildfire dataset contains data on fire activities in Australia starting from 2005. Additional information can be found [here](https://earthdata.nasa.gov/earth-observation-data/near-real-time/firms/c6-mcd14dl).

Variables

- Region: the 7 regions
- Date: in UTC and provide the data for 24 hours ahead
- Estimated_fire_area: daily sum of estimated fire area for presumed vegetation fires with a confidence > 75% for a each region in km2
- Mean_estimated_fire_brightness: daily mean (by flagged fire pixels(=count)) of estimated fire brightness for presumed vegetation fires with a confidence level > 75% in Kelvin
- Mean_estimated_fire_radiative_power: daily mean of estimated radiative power for presumed vegetation fires with a confidence level > 75% for a given region in megawatts
- Mean_confidence: daily mean of confidence for presumed vegetation fires with a confidence level > 75%
- Std_confidence: standard deviation of estimated fire radiative power in megawatts
- Var_confidence: Variance of estimated fire radiative power in megawatts
- Count: daily numbers of pixels for presumed vegetation fires with a confidence level of larger than 75% for a given region
- Replaced: Indicates with an Y whether the data has been replaced with standard quality data when they are available (usually with a 2-3 month lag). Replaced data has a slightly higher quality in terms of locations

The x-axis labels above aren't very informative. Try concatinating year and month into a string.

`df_grouped.index = df_grouped.index.map(lambda x: f"{x[0]}-{x[1]:02d}")`: This line converts the multi-index (Year, Month) into a single string in the format "Year-Month". The `x[0]` represents the Year, and `x[1]` represents the Month. The `:02d` format ensures that the month is displayed with two digits, e.g., "01" for January.

````python
#create a new index combining Year and Month as a string
df_grouped2 = df.groupby(['Year', 'Month'])['Estimated_fire_area'].mean()
df_grouped2.index = df_grouped2.index.map(lambda x: f"{x[0]}-{x[1]:02d}")

#plot with new index
plt.figure(figsize=(12, 6))
df_grouped2.plot()
plt.xlabel('Year-Month')
plt.ylabel('Average Estimated Fire Area (km²)')
plt.title('Estimated Fire Area over Time')
plt.xticks(rotation=45) #rotate axis labels by 45 degrees
plt.grid(True) #add gridlines
plt.tight_layout() #adjust layout so labels and title fit
plt.show()
````

Trying to fix the months starting at progressively higher numbers instead of January '01' each time (on the x-axis)

`df_grouped.reindex([(d.year, d.month) for d in all_months], fill_value=np.nan)`: This ensures that the reindexing uses tuples of (year, month) correctly, avoiding any subscript errors with integers.

`df_grouped.index = pd.to_datetime([f"{year}-{month:02d}-01" for year, month in df_grouped.index])`: Converts the index to a list of strings in the "YYYY-MM-01" format and then converts it to datetime, ensuring consistent datetime objects.

### TASK 1.5: See the percentage on the pie is not looking so good as it is overlaped for Region SA, TA, VI 

remove the autopct fromm pie function and pass the following to plt.legend() after plt.title() <br>
`[(i,round(k/region_counts.sum()*100,2)) for i,k in zip(region_counts.index, region_counts)]`


The line plt.legend `([(i, round(k / region_counts.sum() * 100, 2)) for i, k in zip(region_counts.index, region_counts)])` is creating a custom legend for the pie chart. This custom legend includes both the region names and the percentage of the total count each region represents. Here's a breakdown of what this line does:

1. `zip(region_counts.index, region_counts)`:

* `region_counts.index` provides the region names.
* `region_counts` provides the count values for each region.
* `zip` pairs each region name with its corresponding count value.

2. `[(i, round(k / region_counts.sum() * 100, 2)) for i, k in zip(region_counts.index, region_counts)]`:

* This is a list comprehension that iterates over the pairs from the `zip` object.
* `i` is the region name.
* `k` is the count value for that region.
* `k / region_counts.sum() * 100` calculates the percentage of the total count that this region's count represents.
* `round(..., 2)` rounds this percentage to two decimal places.
* The tuple `(i, round(k / region_counts.sum() * 100, 2))` includes the region name and the rounded percentage.
* The result is a list of tuples, where each tuple contains a region name and its corresponding percentage.

3. plt.legend(...):

* This function creates a legend for the plot.
* The argument `[(i, round(k / region_counts.sum() * 100, 2)) for i, k in zip(region_counts.index, region_counts)]` provides the labels for the legend. Each label is a tuple containing the region name and its percentage of the total.

````python
# To fix the issue in their solution
# Creating a pie chart to visualize the portion of count of pixels for presumed vegetation fires across regions
plt.figure(figsize=(10, 6))
# Grouping the data by region and summing the counts
region_counts = df.groupby('Region')['Count'].sum()
# Creating the pie chart using plt.pie function
# Labels are set to the region names, and autopct is used to display percentage
plt.pie(region_counts, labels=region_counts.index)
plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region')
plt.legend([(i,round(k/region_counts.sum()*100,2)) for i,k in zip(region_counts.index, region_counts)])
plt.axis('equal') #ensures plot is drawn as a circle
plt.show()
````

### TASK 1.9: Let's mark these seven regions on the Map of Australia using Folium
<br> we have created a dataframe for you containing the regions, their latitudes and longitudes. <br> For australia use [-25, 135] as location to create the map


````python
region_data = {'region':['NSW','QL','SA','TA','VI','WA','NT'], 'Lat':[-31.8759835,-22.1646782,-30.5343665,-42.035067,-36.5986096,-25.2303005,-19.491411], 
               'Lon':[147.2869493,144.5844903,135.6301212,146.6366887,144.6780052,121.0187246,132.550964]}
reg=pd.DataFrame(region_data)
reg
````

````python
Australia = folium.Map(location=[-25, 135], zoom_start = 4)
Australia
````

````python
# instantate a feature group for the regions
regions = folium.map.FeatureGroup()

#loop through the reg dataframe to add the lat long of each region to the regions feature group
for lat, lon, in zip(reg.Lat, reg.Lon): # lat and lon are created here, reg.Lat and reg.Lon are columns pulled from the 'reg' dataframe, zip is fxn that creates tuples of two lists
    regions.add_child(
        folium.vector_layers.CircleMarker(
            [lat,lon], #marker position (tuple)
            radius=5,
            color="blue",
            fill=True,
            fill_color="yellow",
            fill_opacity=0.5
            
        )
    )

#Add regions to map
Australia.add_child(regions)
````
